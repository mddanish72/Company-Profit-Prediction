```python
# Creating dummy variables for catgeorical variable 'State'

dataset['NewYork_State']=np.where(dataset['State']=='New York',1,0)
dataset['California_State']=np.where(dataset['State']=='California',1,0)
dataset['Florida_State']=np.where(dataset['State']=='Florida',1,0)

dataset.drop(columns=['State'],axis=1,inplace=True)
dataset
```

```python
dependent_variable='Profit'
independent_variable=dataset.columns.tolist()
independent_variable.remove(dependent_variable)
independent_variable
```

```python
x=dataset[independent_variable].values
y=dataset[dependent_variable].values
```

```python
#Splitting the data into training set and test set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=8103)
```

```python
#Transforming data

from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
x_train[0:5]
```

```python
### Building the model

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# Predicting the result

y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Check score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
### Testing the maximum accuracy of the model

accuracy1=[]
for i in range(0,10000):
    xtrain2,xtest2,ytrain2,ytest2=train_test_split(x,y,train_size=0.8,random_state=i)
    model2=LinearRegression()
    model2.fit(xtrain2,ytrain2)
    ypred1=model2.predict(xtest2)
    r2_2=r2_score(ytest2,ypred1)
    accuracy1.append(r2_2)
```

```python
accuracy1
```

```python
np.max(accuracy1)
```
